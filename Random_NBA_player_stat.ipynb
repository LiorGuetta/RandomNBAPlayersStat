```
def mutation_stat(prob):
  r1 = random.randint(0, 100)
  if r1<prob:
    x = random.randint(8, 12)
    return x/10
  else:
    return 1
```

```
def mutation_sign(sign,prob):
  r1 = random.randint(0, 100)
  if r1<prob:
    if sign == "<":
      r2 = random.uniform(0, 10)
      if r2<9:
        return ">"
      else:
        return "="
    elif sign == ">":
      r3 = random.uniform(0, 10)
      if r3<9:
        return ">"
      else:
        return "="
    elif sign == "=":
      r4 = random.randint(0, 10)
      if r4<5:
        return ">"
      else:
        return "<"
  else:
    return sign
```

```
def five_mutations(comparison,values):
  new_signs_first =[]
  new_stats_first =[]
  new_signs_second =[]
  new_stats_second =[]
  new_signs_third =[]
  new_stats_third =[]
  new_signs_fourth =[]
  new_stats_fourth =[]
  new_signs_fifth =[]
  new_stats_fifth =[]
  chosen_signs = []
  chosen_stats = []
  fit_score_one = 0
  fit_score_two = 0
  fit_score_three = 0
  fit_score_four = 0
  fit_score_five = 0
  total_score = 0
  number_one_first = 0
  number_one_last = 0
  number_two_first = 0
  number_two_last = 0
  number_three_first = 0
  number_three_last = 0
  number_four_first = 0
  number_four_last = 0
  number_five_first = 0
  number_five_last = 0

  for i in range(5):
    new_signs = []
    new_stats = []
    for j in range(8):
      new_signs.append(mutation_sign(comparison[j],15))
      new_stats.append(mutation_stat(15) *values[j])
      players_counter = 0
      players_names = []
    for k in range(len(id_arr)):
      if new_signs[0] == "=":
        if new_stats[0] != points_arr[k]:
          continue
      elif new_signs[0] == ">":
        if points_arr[k] >= new_stats[0]:
          continue
      elif new_signs[0] == "<":
        if points_arr[k] <= new_stats[0]:
          continue
      if new_signs[1] == "=":
        if new_stats[1] != blocks_arr[k]:
          continue
      if new_signs[1] == ">":
        if blocks_arr[k] >= new_stats[1]:
          continue
      if new_signs[1] == "<":
        if blocks_arr[k] <= new_stats[1]:
          continue
      if new_signs[2] == "=":
        if new_stats[2] != rebounds_arr[k]:
          continue
      if new_signs[2] == ">":
        if rebounds_arr[k] >= new_stats[2]:
          continue
      if new_signs[2] == "<":
        if rebounds_arr[k] <= new_stats[2]:
          continue
      if new_signs[3] == "=":
        if new_stats[3] != steals_arr[k]:
          continue
        if new_signs[3] == ">":
          if steals_arr[k] >= new_stats[3]:
            continue
      if new_signs[3] == "<":
        if steals_arr[k] <= new_stats[3]:
          continue
      if new_signs[4] == "=":
        if new_stats[4] != assists_arr[k]:
          continue
      if new_signs[4] == ">":
        if assists_arr[k] >= new_stats[4]:
          continue
      if new_signs[4] == "<":
        if assists_arr[k] <= new_stats[4]:
          continue
      if new_signs[5] == "=":
        if new_stats[5] != fg_percent_arr[k]:
          continue
        if new_signs[5] == ">":
          if fg_percent_arr[k] >= new_stats[5]:
            continue
        if new_signs[5] == "<":
          if fg_percent_arr[k] <= new_stats[5]:
            continue
      if new_signs[6] == "=":
        if new_stats[6] != games_played_arr[k]:
          continue
        if new_signs[6] == ">":
          if games_played_arr[k] >= new_stats[6]:
            continue
        if new_signs[6] == "<":
          if games_played_arr[k] <= new_stats[6]:
            continue
      if new_signs[7] == "=":
        if new_stats[7] != age_arr[k]:
          continue
        if new_signs[7] == ">":
          if age_arr[k] >= new_stats[7]:
            continue
        if new_signs[7] == "<":
          if age_arr[k] <= new_stats[7]:
            continue
      players_counter += 1
      players_names.append(id_arr[k])
    fit_score = 0
    if players_counter == 0:
      fit_score = 0
      if i==0:
        number_one_first = 0
        number_one_last = 0
      elif i==1:
        number_two_first = number_one_last
        number_two_last = number_one_last
      elif i==2:
        number_three_first = number_two_last
        number_three_last = number_two_last
      elif i==3:
        number_four_first = number_three_last
        number_four_last = number_three_last
      elif i==4:
        number_five_first = number_four_last
        number_five_last = number_four_last
    else:
      fit_score = (1/players_counter)*1000
      total_score += fit_score
      if i==0:
        number_one_first = 0
        number_one_last = fit_score
        fit_score_one = fit_score
      elif i==1:
        number_two_first = number_one_last
        number_two_last = fit_score+number_two_first
        fit_score_two = fit_score
      elif i==2:
        number_three_first = number_two_last
        number_three_last = fit_score+number_three_first
        fit_score_three = fit_score
      elif i==3:
        number_four_first = number_three_last
        number_four_last = fit_score+number_four_first
        fit_score_four = fit_score
      elif i==4:
        number_five_first = number_four_last
        number_five_last = fit_score+number_five_first
        fit_score_five = fit_score
    if i==0:
      new_signs_first = new_signs
      new_stats_first = new_stats
    elif i==1:
      new_signs_second = new_signs
      new_stats_second = new_stats
    elif i==2:
      new_signs_third = new_signs
      new_stats_third = new_stats
    elif i==3:
      new_signs_fourth = new_signs
      new_stats_fourth = new_stats
    elif i==4:
      new_signs_fifth = new_signs
      new_stats_fifth = new_stats
  random_number = random.uniform(0, total_score)
  if random_number>number_one_first and random_number<number_one_last:
    return new_signs_first,new_stats_first,fit_score_one
  elif random_number>number_two_first and random_number<number_two_last:
    return new_signs_second,new_stats_second,fit_score_two
  elif random_number>number_three_first and random_number<number_three_last:
    return new_signs_third,new_stats_third,fit_score_three
  elif random_number>number_four_first and random_number<number_four_last:
    return new_signs_fourth,new_stats_fourth,fit_score_four
  elif random_number>number_five_first and random_number<number_five_last:
    return new_signs_fifth,new_stats_fifth,fit_score_five
  else:
    return comparison,values
```
